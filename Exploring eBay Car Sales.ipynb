Looking at the data, it looks as though there are two types of data types within the data set: int64 and objects. Most of these objects are strings, with null entries in gearbox, model, fuelType, and notRepairedDamage. 

The column names within the dataset use CamelCase instead of Python's preferred snake_case. We should fix them to be the preferred format. 

Due to this, it looks as though some clean up of the data is required. 
Let's start by renaming some of the columns:

```python
autos.rename(columns = {'yearOfRegistration': 'registration_year', 'monthOfRegistration': 'registration_month', 'notRepairedDamage': 'unrepaired_damange', 'dateCreated': 'ad_created'})
```

Let's create a function that will make the columns snake_case:
- If the character is lowercase already, append it to the new column name
- If the character is uppercase, append an underscore indicating a space and then append the lowercase version of the character to the column name
- If the character is a special symbol (such as an underscore already), append it normally.

```python
def SnakeCase(colname):
    newcol = ''
    for char in colname:
        if char.islower():
            newcol += char
        elif char.isupper():
            newcol += '_'
            newcol += char.lower()
        else:
            newcol += char
    return (newcol)

#Test Statement
print(SnakeCase('dateCrawled'))
```

Now that we have a function to convert the columns to snake case, let's change all of the columns to snake case: 

```python
newcols = []
for col in autos.columns:
    newcols.append(SnakeCase(col))

autos.columns = newcols

autos.head()
```

Now we'll look at the different brands in each column and aggregate them. Let's start by examining the `brand` column:

```python
autos['brand'].value_counts(normalize = True, dropna = False).sort_values(ascending = False)
```

It looks like German cars are the most popular brand having 6 of the top 10 brands. With Volkswagon being the most popular car listed on eBay having the number of listings as the 2nd and 3rd most popular brands (Opel and BMW) combined.

Based off of the percentages displayed here, let's take the top 16 brands, all with >1% within the dataset. 

```python
top16 = autos['brand'].value_counts(normalize = True,dropna = False).sort_values(ascending = False)[:16]
print(top16)
```

```python
brands = top16.index
print(brands)
```

Now we'll create a dictionary containing each brand and its mean price. After making the dictionary, we'll sort it.

```python
price_dic = {}
for brand in brands:
    mean_price = autos.loc[autos['brand']==brand, 'price_usd'].mean()
    price_dic[brand] = round(mean_price,2)
print(price_dic)
```

```python
sorted_dic = sorted(price_dic.items(), key=lambda x: x[1], reverse = True)    
print(sorted_dic)
```

On average, it looks as though Audi, Mercedes Benz, and BMW are the most expensive brands all with average prices over 8000. The cheapest of the top 16 brands being Renault, Fiat, and Opel all within the 2000 dollar range. 

Finally, we'll create a dataframe with the brand data, including the mean price and mean milage of all of the brands by converting each to series and then combining them.

```python
miles_dic = {}
for brand in brands:
    mean_mile = autos.loc[autos['brand']==brand, 'odometer_km'].mean()
    miles_dic[brand] = round(mean_mile,2)
print(miles_dic)
```

```python
sorted_dic = sorted(miles_dic.items(), key=lambda x: x[1], reverse = True)    
print(sorted_dic)
```

The brands with the highest milege seems to correlate with the most expensive brands, with BMW, Mercedes Benz, and Audi having the highest mileage. However the cheapest brands do not have the least mileage. 

```python
prices = pd.Series(price_dic)
miles = pd.Series(miles_dic)
df = pd.DataFrame(prices, columns=['mean_price'])
df['mean_mileage'] = miles
print(df)
```
